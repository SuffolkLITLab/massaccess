```python
def download_pdf(url,name):
    try:
        r = requests.get(url) # create HTTP response object 
        with open("core_assets/pdfs/%s"%name,'wb') as f: 
            f.write(r.content) 
        #print("\nWrote PDF from %s\n"%url)
        return True
    except:
        print("\nERROR WRITING PDF!!! from %s\n"%url)
        return False
```

```python
def header(page,lang):
    title = df_static.loc['title'][lan[lang]]
    desc = df_static.loc['desc'][lan[lang]]
    homepage = df_static.loc['homepage'][lan[lang]]
    ma_icon = df_static.loc['ma_icon'][lan[lang]]
    language = df_static.loc['language'][lan[lang]]
    languages = df_static.loc['languages'][lan[lang]]
    about_us  = df_static.loc['about_us'][lan[lang]]
    news  = df_static.loc['news'][lan[lang]]
    my_forms = df_static.loc['my_forms'][lan[lang]]
    alert_test = df_static.loc['alert_test'][lan[lang]]
    alert_1 = df_static.loc['alert_1'][lan[lang]]
    alert_2 = df_static.loc['alert_2'][lan[lang]]
    alert_3 = df_static.loc['alert_3'][lan[lang]]

    if (lang=="en"):
        suff = ""
    else:
        suff = lang

    if (page != "home"):
        path = "../"
        if (lang =="en"):
            en_link = "./"
            es_link = "es"
        elif (lang =="es"):
            en_link = "../"            
            es_link = "./"
    else:
        path = ""
        if (lang =="en"):
            en_link = "./"
            es_link = "es"
        elif (lang =="es"):
            en_link = "../"            
            es_link = "./"
    
    if (lang !="en"):
        path = path + "../"
    
    html = """<!DOCTYPE html>
<HTML xmlns="http://www.w3.org/1999/xhtml"
      xmlns:og="http://ogp.me/ns#"
      xmlns:fb="http://www.facebook.com/2008/fbml"
      lang="%s">
<HEAD>
	<title>%s</title>
	<meta http-equiv="Content-type" content="text/html;charset=UTF-8"/>
	<meta name="viewport" content="width=device-width, initial-scale=1.0" />
	<meta name="apple-mobile-web-app-capable" content="no" />
    <meta name="description" content="%s"/>
	<link rel="apple-touch-icon" href="%simages/icon_300.png"/>
	<meta property="og:type" content="website"/>
	<meta property="og:title" content="%s"/>
	<meta property="og:description" content="%s"/>
	<meta property="og:image" content="http://massaccess.suffolklitlab.org.s3-website-us-east-1.amazonaws.com/images/screenshot_2000x1000.png"/>

	<meta name="twitter:card" content="summary_large_image">
	<meta name="twitter:site" content="@SuffolkLITLab">
	<meta name="twitter:creator" content="@SuffolkLITLab">
	<meta name="twitter:title" content="%s">
	<meta name="twitter:description" content="%s">
	<meta name="twitter:image" content="http://massaccess.suffolklitlab.org.s3-website-us-east-1.amazonaws.com/images/screenshot_2000x1000.png"/>

  <link href='https://fonts.googleapis.com/css?family=Inter' rel='stylesheet'>
  <link rel="prenconnect" href="https://google-analytics.com">
  <link rel="stylesheet" type="text/css" href="%scss/style.css?v=%s">

  <!-- Global site tag (gtag.js) - Google Analytics -->
  <script async src="https://www.googletagmanager.com/gtag/js?id=UA-163204491-1"></script>
  <script>
    window.dataLayer = window.dataLayer || [];
    function gtag(){dataLayer.push(arguments);}
    gtag('js', new Date());

    gtag('config', 'UA-163204491-1');
  </script>
  <script defer src="https://code.jquery.com/jquery-3.6.0.min.js"></script>
  <script defer src="https://code.jquery.com/ui/1.13.0/jquery-ui.min.js"></script>
  <script src="%sjs_bin/scripts.js?v=%s"></script>
"""%(lang,title,desc,path,title,desc,title,desc,path,release,path,release)

    if page=="home" and lang=="en":
        html = html + """
  <script defer src="%sjs_bin/Countable.js?v=%s"></script>
  <script defer src="%sjs_bin/test_train.js?v=%s"></script>
  """%(path,release,path,release)
    
    html = html + f"""
</HEAD>
<BODY BGCOLOR="#ffffff" BACKGROUND="" MARGINWIDTH="0" MARGINHEIGHT="0" onLoad="test_test(`{alert_test}`)">

<div class="header" role="banner">
  <div class="nav_bar">
    <div class="menu">
      <div class="dropdown">
          <div class="nav_menu"></div>
          <div class="nav_menu"></div>
          <div class="nav_menu" style="margin-bottom:17px;"></div>
          <div class="dropdown-content" style="right:-10px;">
            <a href="{path}myforms/{suff}" class="whitelink"><p class="menu_item">{my_forms}</p></a>
            <a href="{path}about/{suff}" class="whitelink"><p class="menu_item">{about_us}</p></a>
            <a href="{path}news/{suff}" class="whitelink"><p class="menu_item">{news}</p></a>
            <!-- --><p class="menu_item_no_hov">{language}:</p>
            <a href="{en_link}" class="whitelink"><p class="menu_item" style="padding-left:25px;"> English</p></a>
            <a href="{es_link}" class="whitelink"><p class="menu_item" style="padding-left:25px;"> Español</p></a>
            <!-- <a href="pt" class="whitelink"><p class="menu_item" style="padding-left:25px;"> Portuguese</p></a>
            <a href="pt-cv" class="whitelink"><p class="menu_item" style="padding-left:25px;"> Cape Verdean</p></a>
            <a href="ht" class="whitelink"><p class="menu_item" style="padding-left:25px;"> Haitian-Creole</p></a>
            <a href="ar" class="whitelink"><p class="menu_item" style="padding-left:25px;"> Arabic</p></a>
            <a href="zh" class="whitelink"><p class="menu_item" style="padding-left:25px;"> Chinese</p></a>
            <a href="km" class="whitelink"><p class="menu_item" style="padding-left:25px;"> Khmer</p></a>
            <a href="ru" class="whitelink"><p class="menu_item" style="padding-left:25px;"> Russian</p></a>
            <a href="vi" class="whitelink"><p class="menu_item" style="padding-left:25px;"> Vietnamese</p></a>-->
          </div>
        </div>
      </div>
    <a href="{path}{suff}" alt="{homepage}" tabindex="1"><img id="lit_logo" src="{path}images/lit_logo_light.png" alt="{ma_icon}" class="lit_logo">
    <div class="title" style="line-height:20px;margin-top:9px">
    <span style="font-size:14px;">Court Forms Online</span><br/>
    MassAccess</div></a>
    <div class="right_nav_container">
      <div class="right_nav"><a href="{path}myforms/{suff}" tabindex="6">{my_forms}</a></div>
      <div class="right_nav"><a href="{path}about/{suff}" tabindex="5">{about_us}</a></div>
      <div class="right_nav"><a href="{path}news/{suff}" tabindex="4">{news}</a></div>
      <div class="right_nav">
        <!-- --><div class="dropdown">
          {languages}
          <div class="arrow-down"></div>
          <div class="dropdown-content" style="left:-10px;">
            <a href="{en_link}" class="whitelink" tabindex="2"><p class="menu_item">English</p></a>
            <a href="{es_link}" class="whitelink" tabindex="3"><p class="menu_item">Español</p></a>
            <!-- <a href="pt" class="whitelink"><p class="menu_item">Portuguese</p></a>
            <a href="pt-cv" class="whitelink"><p class="menu_item">Cape Verdean</p></a>
            <a href="ht" class="whitelink"><p class="menu_item">Haitian-Creole</p></a>
            <a href="ar" class="whitelink"><p class="menu_item">Arabic</p></a>
            <a href="zh" class="whitelink"><p class="menu_item">Chinese</p></a>
            <a href="km" class="whitelink"><p class="menu_item">Khmer</p></a>
            <a href="ru" class="whitelink"><p class="menu_item">Russian</p></a>
            <a href="vi" class="whitelink"><p class="menu_item">Vietnamese</p></a> -->
          </div>
        </div><!-- -->
      </div>
    </div>
  </div>
</div>"""
    
    html = html + """<div id="alert_test"></div>"""
    
    html = html #+ alert_3    

    if (page=="helpline"):
        html = html + alert_2
    elif (page != "about" and page != "privacy"):
        html = html #+ alert_1
       
    return html

def topics(page,lang):
    message = df_static.loc['message'][lan[lang]]
    housing_alt = df_static.loc['housing_alt'][lan[lang]]
    housing = df_static.loc['housing'][lan[lang]]
    dv_alt = df_static.loc['dv_alt'][lan[lang]]
    dv = df_static.loc['dv'][lan[lang]]
    guardianship_alt = df_static.loc['guardianship_alt'][lan[lang]]
    guardianship = df_static.loc['guardianship'][lan[lang]]
    family_alt = df_static.loc['family_alt'][lan[lang]]
    family = df_static.loc['family'][lan[lang]]
    employment_alt = df_static.loc['employment_alt'][lan[lang]]
    employment = df_static.loc['employment'][lan[lang]]
    consumer_alt = df_static.loc['consumer_alt'][lan[lang]]
    consumer = df_static.loc['consumer'][lan[lang]]
    health_alt = df_static.loc['health_alt'][lan[lang]]
    health = df_static.loc['health'][lan[lang]]
    income_alt = df_static.loc['income_alt'][lan[lang]]
    income = df_static.loc['income'][lan[lang]]
    immigration_alt = df_static.loc['immigration_alt'][lan[lang]]
    immigration = df_static.loc['immigration'][lan[lang]]
    education_alt = df_static.loc['education_alt'][lan[lang]]
    education = df_static.loc['education'][lan[lang]]
    appeals_alt = df_static.loc['appeals_alt'][lan[lang]]
    appeals = df_static.loc['appeals'][lan[lang]]
    safety_alt = df_static.loc['safety_alt'][lan[lang]]
    safety = df_static.loc['safety'][lan[lang]]
    other_alt = df_static.loc['other_alt'][lan[lang]]
    other = df_static.loc['other'][lan[lang]]
    hide_cat = df_static.loc['hide_cat'][lan[lang]]
    show_cat = df_static.loc['show_cat'][lan[lang]]

    housing_suff = gray_or_not("housing",lang)  
    dv_suff = gray_or_not("dv",lang)  
    guardianship_suff = gray_or_not("guardianship",lang)  
    family_suff = gray_or_not("family",lang)  
    employment_suff = gray_or_not("employment",lang)  
    consumer_suff = gray_or_not("consumer",lang)  
    health_suff = gray_or_not("health",lang)  
    income_suff = gray_or_not("income",lang)  
    immigration_suff = gray_or_not("immigration",lang)  
    education_suff = gray_or_not("education",lang)  
    appeals_suff = gray_or_not("appeals",lang)  
    safety_suff = gray_or_not("safety",lang)  
    other_suff = gray_or_not("other",lang)  
    
    housing_style = ""
    dv_style = ""
    guardianship_style = ""
    family_style = ""
    employment_style = ""
    consumer_style = ""
    health_style = ""
    income_style = ""
    immigration_style = ""
    education_style = ""
    appeals_style = ""
    safety_style = ""
    other_style = ""    
    
    if (page == "housing"):
        housing_style = "style=\"border: 2px solid #0072E8\""
    if (page == "dv"):
        dv_style = "style=\"border: 2px solid #0072E8\""
    if (page == "guardianship"):
        guardianship_style = "style=\"border: 2px solid #0072E8\""
    if (page == "family"):
        family_style = "style=\"border: 2px solid #0072E8\""
    if (page == "employment"):
        employment_style = "style=\"border: 2px solid #0072E8\""
    if (page == "consumer"):
        consumer_style = "style=\"border: 2px solid #0072E8\""
    if (page == "health"):
        health_style = "style=\"border: 2px solid #0072E8\""
    if (page == "income"):
        income_style = "style=\"border: 2px solid #0072E8\""
    if (page == "immigration"):
        immigration_style = "style=\"border: 2px solid #0072E8\""
    if (page == "education"):
        education_style = "style=\"border: 2px solid #0072E8\""
    if (page == "appeals"):
        appeals_style = "style=\"border: 2px solid #0072E8\""
    if (page == "safety"):
        safety_style = "style=\"border: 2px solid #0072E8\""
    if (page == "other"):
        other_style = "style=\"border: 2px solid #0072E8\""
        
    if (lang=="en"):
        suff = ""
    else:
        suff = lang
        
    if (page != "home"):
        path = "../"
    else:
        path = ""

    if (lang !="en"):
        path = path + "../"
    
    hide_width = "170px"
    show_width = "175px"
    if (lang == "es"):
        hide_width = "200px"
        show_width = "195px"        
        
    if (page == "home" and lang =="en"): 
        html = """
<div class="divTable" style="background:#e6ecfc">
<div class="divRow">
<div class="divCell_desc">

  <div class="spot_description">
    <h1>Find and file court forms online during the emergency</h1>
     <p>
     This site offers a way for you to reach the court during the COVID-19 crisis. It provides court forms and self-help materials for areas of urgent legal need. Describe your issue here, and our issue spotter will look for helpful resources. You can also browse forms below by category.
     </p> 
     <!--<p>Most forms are Massachusetts specific, but a few, like the <a href="https://apps.suffolklitlab.org/start/cdc_moratorium" target="_blank">CDC Eviction Moratorium Assistant</a>, can be used across the United States.</p>--> 
  </div>

</div>

<div class="divCell_input" style="background:#fff">

<div class="spot_input">

    <div style="display:none;float:left;background:#eee;width:100%;margin:0px 0 20px 0;padding:0 15px;box-sizing: border-box;border: solid 1px #555;">
    <p>
      Token: <input id='bearer' name='bearer' style="width:300px;" value="516ca8844a07ccbd380cbf9b4255dc7d0449a8805b55fdd525743eea31bfb054">&nbsp;&nbsp;
      Cutoffs:&nbsp;&nbsp; Upper: <input type=number id='cutoff-upper' name='cutoff-upper' style="width:45px;" value="0.5">&nbsp;&nbsp;
      Pred: <input type=number id='cutoff-pred' name='cutoff-pred' style="width:45px;" value="0.5">&nbsp;&nbsp;
      Lower: <input type=number id='cutoff-lower' name='cutoff-lower' style="width:45px;" value="0.25">
    </p>
    </div>

    <label for="text"><h2 style="text-align:center;margin:25px 0px 0px 0px;text-align:left;">What seems to be the problem?
    <div style="float:right;margin:3px 0 4px 0;font-weight: normal;">
    <span id="word_count" name="word_count" class="subtitle" style="font-size:14px;float:right;color:#bd5757;">&nbsp;&nbsp;<span id="result__words">0</span> words</span>
    </div></h2></label>

    <textarea name='text' id="text" style="width:100%;height:150px;font-size:15px;padding:10px;box-sizing: border-box; border-radius: 15px; border: solid 1px #5e5e5e;resize:none" placeholder="Share your story here"></textarea>

    <div style="float:right;margin:8px 0 15px 0;font-weight: normal;">
      <span class="subtitle" style="color:#6f6f6f;font-size:14px;"> Reply should be at least 5 words long. Longer replies get better results. Do <b>NOT</b> include identifying information like your name or address.
      </span>
    </div>

<table><tr><td width="70%">

  <table><tr><td valign="top">
  <input id="save-id" name="save-id" type="checkbox" CHECKED style="height:20px;width:20px;">
  </td><td>
    <span class="subtitle"><label for="save-id"> Use my reply to help people facing similar issues. <a href="https://spot.suffolklitlab.org/data/user-driven/" rel="noopener" target="_blank" class="litelink">Learn how</a></label></span>
  </td></tr></table>

    <input type="hidden" id="man_jur" value="MA"/>
    <input type="hidden" id="auto_jur" value="MA"/>

</td><td width="30%">

<a href="javascript:void('')" onClick="make_query();gtag('event', 'Spot', {'event_category':'API','event_label':'send to Spot'});"><div class="find">Find Help</div></a>

</td></tr></table>

</div>

</div></div></div>



<div class="white_band" role="complementary">
<div class="content">
  <div id="output">

  <div id="loading" style="display:none;width:100%;margin:50px auto;text-align:center;">
  <img src="images/Ajax-loader.gif" width="35px;" height="35px;"/>
  </div>

  <div id="response" style="float:left;display:none;width:100%;margin:20px auto;">
  </div>

</div>
</div>
</div>


</div>
    <div class="gray_band_w_text" role="main">
    <div class="content">
      <center><h2>Browse Court Forms by Category</h2></center>
      <div class="buttons_cats_wide">
"""
    elif (page == "home" and lang !="en"):
        html = """
    <div class="gray_band" role="main">
    <div class="content">
    <div class="description">%s
    </div>
    <div class="buttons">"""%message        
    else: 
        html = """  
    <div class="gray_band" style="padding:0px;" role="navigation">
    <div id="cat_buttons" style="display:none;float:left;margin-bottom:8px;width:100%;">
      <div class="buttons_cats">"""
        
    
    if (page == "home" and lang =="en"):
        gtag = "onClick=\"gtag('event', 'navigation', {'event_category':'pick topic','event_label':'from list'});\""
    else:
        gtag = ""   
        
        
    html = html + """
      <a href="%shousing/%s" class="abutton%s" %s><div class="button" %s>
        <div class="button_cell_l">
          <img src="%simages/icon_housing%s.png" alt="%s" class="topic_icon">
        </div>
        <div class="button_cell">
          <span class="b_text">%s</span>
        </div>
      </div></a>

      <a href="%sdv/%s" class="abutton%s" %s><div class="button" %s>
        <div class="button_cell_l">
          <img src="%simages/icon_domestic_violence%s.png" alt="%s" class="topic_icon">
        </div>
        <div class="button_cell">
          <span class="b_text">%s</span>
        </div>
      </div></a>

      <a href="%sappeals/%s" class="abutton%s" %s><div class="button" %s>
        <div class="button_cell_l">
          <img src="%simages/icon_appeals%s.png" alt="%s" class="topic_icon">
        </div>
        <div class="button_cell">
          <span class="b_text">%s</span>
        </div>
      </div></a>

      <a href="%sfamily/%s" class="abutton%s" %s><div class="button" %s>
        <div class="button_cell_l">
          <img src="%simages/icon_family%s.png"  alt="%s" class="topic_icon">
        </div>
        <div class="button_cell">
          <span class="b_text">%s</span>
        </div>
      </div></a>
      
      <a href="%sguardianship/%s" class="abutton%s" %s><div class="button" %s>
        <div class="button_cell_l">
            <img src="%simages/icon_guardianship%s.png" alt="%s" class="topic_icon">
        </div>
        <div class="button_cell">
            <span class="b_text">%s</span>
        </div>
      </div></a>

      <a href="%semployment/%s" class="abutton%s" %s><div class="button" %s>
        <div class="button_cell_l">
          <img src="%simages/icon_employment%s.png" alt="%s" class="topic_icon">
        </div>
        <div class="button_cell">
          <span class="b_text">%s</span>
        </div>
      </div></a>

      <a href="%sconsumer/%s" class="abutton%s" %s><div class="button" %s>
        <div class="button_cell_l">
          <img src="%simages/icon_consumer%s.png" alt="%s" class="topic_icon">
        </div>
        <div class="button_cell">
          <span class="b_text">%s</span>
        </div>
      </div></a>

      <a href="%shealth/%s" class="abutton%s" %s><div class="button" %s>
        <div class="button_cell_l">
          <img src="%simages/icon_health%s.png" alt="%s" class="topic_icon">
        </div>
        <div class="button_cell">
          <span class="b_text">%s</span>
        </div>
      </div></a>

      <a href="%sincome/%s" class="abutton%s" %s><div class="button" %s>
        <div class="button_cell_l">
          <img src="%simages/icon_income%s.png" alt="%s" class="topic_icon">
        </div>
        <div class="button_cell">
          <span class="b_text">%s</span>
        </div>
      </div></a>

      <a href="%simmigration/%s" class="abutton%s" %s><div class="button" %s>
        <div class="button_cell_l">
          <img src="%simages/icon_immigration%s.png" alt="%s" class="topic_icon">
        </div>
        <div class="button_cell">
          <span class="b_text">%s</span>
        </div>
      </div></a>

      <a href="%ssafety/%s" class="abutton%s" %s><div class="button" %s>
        <div class="button_cell_l">
          <img src="%simages/icon_safety_others%s.png" alt="%s" class="topic_icon">
        </div>
        <div class="button_cell">
          <span class="b_text">%s</span>
        </div>
      </div></a>

      <a href="%sother/%s" class="abutton%s" %s><div class="button" %s>
        <div class="button_cell_l">
          <img src="%simages/icon_other%s.png" alt="%s" class="topic_icon">
        </div>
        <div class="button_cell">
          <span class="b_text">%s</span>
        </div>
      </div></a>

    </div>

  </div>"""%(path,suff,housing_suff,gtag,housing_style,path,housing_suff,housing_alt,housing,
           path,suff,dv_suff,gtag,dv_style,path,dv_suff,dv_alt,dv,
           path,suff,appeals_suff,gtag,appeals_style,path,appeals_suff,appeals_alt,appeals,
           path,suff,family_suff,gtag,family_style,path,family_suff,family_alt,family,
           path,suff,guardianship_suff,gtag,guardianship_style,path,guardianship_suff,guardianship_alt,guardianship,
           path,suff,employment_suff,gtag,employment_style,path,employment_suff,employment_alt,employment,
           path,suff,consumer_suff,gtag,consumer_style,path,consumer_suff,consumer_alt,consumer,
           path,suff,health_suff,gtag,health_style,path,health_suff,health_alt,health,
           path,suff,income_suff,gtag,income_style,path,income_suff,income_alt,income,
           path,suff,immigration_suff,gtag,immigration_style,path,immigration_suff,immigration_alt,immigration,
           path,suff,safety_suff,gtag,safety_style,path,safety_suff,safety_alt,safety,
           path,suff,other_suff,gtag,other_style,path,other_suff,other_alt,other)
    
    if (page != "home"):
        html = html + """    <a href="#" onClick="$('#hcats').hide();$('#cat_buttons').hide();$('#scats').show();" class="litelink"><div id="hcats" class="cats_link" style="display:none;width:%s"><div class="show_table"><div class="show_cell">%s</div><div class="show_cell" style="width:1%%"><div class="arrow-up"></div></div></div></div></a>

    <a href="javascript:void('');" onClick="$('#scats').hide();$('#cat_buttons').show();$('#hcats').show();" class="litelink"><div id="scats" class="cats_link" style="width:%s"><div class="show_table"><div class="show_cell">%s</div><div class="show_cell" style="width:1%%"><div class="arrow-down_cat"></div></div></div></div></a>

  </div>"""%(hide_width,hide_cat,show_width,show_cat)
    
    html = html + "</div>"
    
    return html

def logos(page,lang):
    lit_logo = df_static.loc['lit_logo'][lan[lang]]
    suffolk_law = df_static.loc['suffolk_law'][lan[lang]]
    a2j_logo = df_static.loc['a2j_logo'][lan[lang]]
    docassemble_logo = df_static.loc['docassemble_logo'][lan[lang]]
    gbls_logo = df_static.loc['gbls_logo'][lan[lang]]
    nela_logo = df_static.loc['nela_logo'][lan[lang]]
    mlri_logo = df_static.loc['mlri_logo'][lan[lang]]
    tp_logo = df_static.loc['tp_logo'][lan[lang]]
    appgeo_logo = df_static.loc['appgeo_logo'][lan[lang]]
    nulawlab_logo = df_static.loc['nulawlab_logo'][lan[lang]]
    cl_logo = df_static.loc['cl_logo'][lan[lang]]
    documate_logo = df_static.loc['documate_logo'][lan[lang]]
    clavic_logo = df_static.loc['clavic_logo'][lan[lang]]

    if (lang=="en"):
        suff = ""
    else:
        suff = lang

    if (page != "home"):
        path = "../"
    else:
        path = ""

    if (lang !="en"):
        path = path + "../"
        
    html = """
      <div class="content_bar_logos_1">
        <div class="logo_table">
          <div class="logo_cell" style="height:60px;">
            <!-- TODO(brycew): add width attributes to these logos, based on the aspect ratio of the images. -->
            <a href="https://suffolklitlab.org/" rel="noopener" target="_blank"><img src="%simages/lit_logo.png" height="50px" width="58px" alt="%s" class="logo"></a>
          </div>
        </div>

        <div class="logo_table">
          <div class="logo_cell" style="height:60px;">
            <a href="https://www.suffolk.edu/law" rel="noopener" target="_blank"><img src="%simages/suffolk_law.png" height="45px" width="320px" alt="%s" class="law_logo"></a>
          </div>
        </div>

        <div class="logo_table">
          <div class="logo_cell" style="height:60px;">
            <a href="http://www.massa2j.org/a2j/" rel="noreferrer" target="_blank"><img src="%simages/a2j_logo.png" height="35px" width="141px" alt="%s" class="logo"></a>
          </div>
        </div>
        
        <div class="logo_table">
          <div class="logo_cell" style="height:60px;">
            <a href="https://docassemble.org/" rel="noreferrer" target="_blank"><img src="%simages/docassemble_logo.png" height="60px" width="131px" alt="%s" class="logo"></a>
          </div>
        </div>
      </div>

      <div class="content_bar_logos_2">
        <div class="logo_table">
          <div class="logo_cell" style="height:100px;">
            <a href="https://massclavc.org/" rel="noreferrer" target="_blank"><img src="%simages/clavic_logo.png" height="60px" width="91px" alt="%s" class="logo"></a>
          </div>
        </div>

        <div class="logo_table">
          <div class="logo_cell" style="height:100px;">
            <a href="https://www.gbls.org/" rel="noreferrer" target="_blank"><img src="%simages/gbls_logo.png" height="60px" width="163px" alt="%s" class="logo"></a>
          </div>
        </div>

        <div class="logo_table">
          <div class="logo_cell" style="height:100px;">
            <a href="https://www.northeastlegalaid.org/" rel="noreferrer" target="_blank"><img src="%simages/ne_legal_aid_logo.png" height="60px" width="90px" alt="%s" class="logo"></a>
          </div>
        </div>

        <div class="logo_table">
          <div class="logo_cell" style="height:100px;">
            <a href="https://www.mlri.org/" rel="noreferrer" target="_blank"><img src="%simages/mlri_logo.png" height="40px" width="192px" alt="%s" class="logo"></a>
          </div>
        </div>
        
        <div class="logo_table">
          <div class="logo_cell" style="height:100px;">
            <a href="https://www.theoryandprinciple.com/" rel="noreferrer" target="_blank"><img src="%simages/tp_logo.png" height="40px" width="141px" alt="%s" class="logo"></a>
          </div>
        </div>
      </div>

      <div class="content_bar_logos_3">
        <div class="logo_table">
          <div class="logo_cell" style="height:100px;">
            <a href="https://www.nulawlab.org/" rel="noreferrer" target="_blank"><img src="%simages/nulawlab_logo.png" height="35px" width="152px" alt="%s" class="logo"></a>
          </div>
        </div>

        <div class="logo_table">
          <div class="logo_cell" style="height:100px;">
            <a href="https://www.appgeo.com/" rel="noreferrer" target="_blank"><img src="%simages/appgeo_logo.png" height="45px" width="106px" alt="%s" class="logo"></a>
          </div>
        </div>
        
        <div class="logo_table">
          <div class="logo_cell" style="height:100px;">
            <a href="https://www.bentley.edu/academics/graduate-programs/masters-human-factors/" rel="noreferrer" target="_blank"><img src="%simages/court_UX_logo.png" height="60px" width="147px" alt="%s" class="logo"></a>
          </div>
        </div>
        
        <div class="logo_table">
          <div class="logo_cell" style="height:100px;">
            <a href="https://community.lawyer/" rel="noreferrer" target="_blank"><img src="%simages/cl_logo.png" height="40px" width="137px" alt="%s" class="logo"></a>
          </div>
        </div>
        
        <div class="logo_table">
          <div class="logo_cell" style="height:100px;">
            <a href="https://www.documate.org/" rel="noreferrer" target="_blank"><img src="%simages/documate_logo.png" height="20px" width="154px" alt="%s" class="logo"></a>
          </div>
        </div>
      </div>"""%(path,lit_logo,path,suffolk_law,path,a2j_logo,path,docassemble_logo,path,clavic_logo,path,gbls_logo,path,nela_logo,path,mlri_logo,path,tp_logo,path,nulawlab_logo,path,appgeo_logo,path,appgeo_logo,path,cl_logo,path,documate_logo)
    return html

def footer(page,lang):
    tos = df_static.loc['tos'][lan[lang]]
    contact = df_static.loc['contact'][lan[lang]]
    
    if (lang=="en"):
        suff = ""
    else:
        suff = lang
    
    if (page != "home"):
        path = "../"
    else:
        path = ""

    if (lang !="en"):
        path = path + "../"
        
    html = """<div class="white_band" style="margin-bottom:200px;" role="contentinfo">
  <div class="content_bar">
    <div class="content_bar_in" style="border-bottom:0px solid">
    <center><a href="%sprivacy/%s" class="litelink">%s</a> &nbsp;|&nbsp; <a href="mailto:massaccess@suffolk.edu" class="litelink">%s</a></center>
    </div>
    </div>
</div>

</BODY>

"""%(path,suff,tos,contact)
    
    if page=="home" and lang == "en":
        html = html + """        
<script>
    document.addEventListener('DOMContentLoaded', ()=> {
      $('#text').on('keypress', function(e){
      if(e.keyCode == 13 && !e.shiftKey) {
         e.preventDefault();
         make_query();
         gtag('event', 'Spot', {'event_category':'API','event_label':'find labels'});
        }
      });
    
    // http://sacha.me/Countable/docs/

    var area = document.getElementById('text');

    function callback (counter) {
      //console.log(counter);
      $('#result__words').html(counter.words);
      if ($('#result__words').html()<5) {
        $('#word_count').css('color', '#bd5757');
        //$('#word_count').css('font-weight', 'bold');
      } else if (($('#result__words').html()>=5) && ($('#result__words').html()<150)) {
        $('#text').css('background-color', '');
        $('#word_count').css('color', '#6f6f6f');
        //$('#word_count').css('font-weight', 'normal');
      } else if (($('#result__words').html()>=150) && ($('#result__words').html()<=250)) {
        $('#word_count').css('color', '#387B31');
        //$('#word_count').css('font-weight', 'bold');
      } else if ($('#result__words').html()>250) {
        $('#word_count').css('color', '#6e6f6f');
        //$('#word_count').css('font-weight', 'normal');
      }
    }

    Countable.live(area, callback);
  });
</script>"""
    
    html = html + "</HTML>"
    
    
    return html

#
#
# --------------------------------------------------------------------------------
#
#

def home(lang):
    page = "home"

    how = df_static.loc['how'][lan[lang]]
    step_1_alt = df_static.loc['step_1_alt'][lan[lang]]
    step_1 = df_static.loc['step_1'][lan[lang]]
    step_2_alt = df_static.loc['step_2_alt'][lan[lang]]
    step_2 = df_static.loc['step_2'][lan[lang]]
    step_3_alt = df_static.loc['step_3_alt'][lan[lang]]
    step_3 = df_static.loc['step_3'][lan[lang]]
    about = df_static.loc['about'][lan[lang]]
    developed = df_static.loc['developed'][lan[lang]]
    court_logo = df_static.loc['court_logo'][lan[lang]]

    if (lang=="en"):
        suff = ""
    else:
        suff = lang
    path = ""
    if (lang !="en"):
        path = path + "../"
        
    step_3 = step_3%(path,suff)
    about = about%(path,suff)

    html = header(page,lang) + topics(page,lang)    
    html = html + """
<div class="white_band" role="complementary">
  <div class="content_bar" style="padding:0px;">
    <div class="content_bar_in">
      <h2>%s</h2>
      <div class="steps">
        <div class="step">
          <div class="step_cell">
            <img src="%simages/icon_step_1.png" alt="%s" class="step_icon">
          </div>
          <div class="step_cell">%s
          </div>
        </div>

        <div class="step_mid">
          <div class="step_cell">
            <img src="%simages/icon_step_2.png" alt="%s" class="step_icon">
          </div>
          <div class="step_cell">%s
          </div>
        </div>

        <div class="step">
          <div class="step_cell">
            <img src="%simages/icon_step_3.png" alt="%s" class="step_icon">
          </div>
          <div class="step_cell">%s
          </div>
        </div>
      </div>
    </div>
  </div>
</div>

<div class="white_band" role="complementary">
  <div class="content_bar">
    <div class="content_bar_in">%s
"""%(how,path,step_1_alt,step_1,path,step_2_alt,step_2,path,step_3_alt,step_3,about)

    html = html + logos(page,lang) +"""
    </div>
  </div>
</div>

<div class="white_band" role="complementary">
  <div class="content_bar">
    <div class="content_bar_in">%s

    <div class="content_bar_logos">
        <a href="https://www.mass.gov/orgs/massachusetts-court-system" target="_blank"><img src="%simages/court_seal.png" height="90px" width="90px" alt="%s" class="logo"></a>
    </div>

    </div>
  </div>
</div>"""%(developed,path,court_logo)
    html = html + footer(page,lang)
    return html



def hackathon(lang):
    page="hackathon"
    
    news_header = df_static.loc['hackathon'][lan[lang]]
    news_body = df_static.loc['hackathon_body']["English"]
        
    if (lang=="en"):
        suff = ""
    else:
        suff = lang
    path = "../"
    if (lang !="en"):
        path = path + "../"
        
    #priv_body = news_body%(path,suff)
    
    if (lang !="en"):
        news_body = df_static.loc['news_lang'][lan[lang]] +  news_body

    html = header(page,lang) + topics(page,lang)
    html = html + """  <div class="white_band">
    <div class="content_bar">
      <div class="content_bar_in_text" style="padding:10px;">
        <center><h1>%s</h1></center>
    </div>
  </div>
  </div>

<div class="white_band" role="main">
  <div class="content_bar">
    <div class="content_bar_in_text">
      %s
    </div>
  </div>
</div>"""%(news_header,news_body)
    html = html + footer(page,lang)
    
    return html


def news(lang):
    page="news"
    
    news_header = df_static.loc['news'][lan[lang]]
    news_body = df_static.loc['news_body']["English"]
        
    if (lang=="en"):
        suff = ""
    else:
        suff = lang
    path = "../"
    if (lang !="en"):
        path = path + "../"
        
    #priv_body = news_body%(path,suff)
    
    if (lang !="en"):
        news_body = df_static.loc['news_lang'][lan[lang]] +  news_body

    html = header(page,lang) + topics(page,lang)
    html = html + """  <div class="white_band">
    <div class="content_bar">
      <div class="content_bar_in_text" style="padding:10px;">
        <center><h1>%s</h1></center>
    </div>
  </div>
  </div>

<div class="white_band" role="main">
  <div class="content_bar">
    <div class="content_bar_in_text">
      %s
    </div>
  </div>
</div>"""%(news_header,news_body)
    html = html + footer(page,lang)
    
    return html



def about(lang):
    page="about"
    
    about_header = df_static.loc['about_header'][lan[lang]]
    about_alt = df_static.loc['about_alt'][lan[lang]]
    about_caption= df_static.loc['about_caption'][lan[lang]]
    about_section_1 = df_static.loc['about_section_1'][lan[lang]]
    about_thankyou = df_static.loc['about_thankyou'][lan[lang]]
    about_section_2 = df_static.loc['about_section_2'][lan[lang]]
    developed = df_static.loc['developed'][lan[lang]]
    court_logo = df_static.loc['court_logo'][lan[lang]]
    
    if (lang=="en"):
        suff = ""
    else:
        suff = lang
    path = "../"
    if (lang !="en"):
        path = path + "../"

    html = header(page,lang) + topics(page,lang)        
    html = html + """<div class="white_band">
  <div class="content_bar">
    <div class="content_bar_in_text" style="padding:10px;">
      <center><h1>%s</h1></center>
  </div>
</div>
</div>

<div class="white_band" role="main">
  <div class="content_bar">
    <div class="content_bar_in_text">
    <div class="r_img_embed" style="max-width:450px;">
      <img src="%simages/video_call.png" alt="%s" width="100%%"/>
      <div class="caption">%s</div>
    </div>
    %s
  </div>
</div>

<div class="white_band" role="complementary">
  <div class="content_bar">
    <div class="content_bar_in">
      <h2>%s</h2>
      <p class="p_bar">
        %s
    </p>"""%(about_header,path,about_alt,about_caption,about_section_1,about_thankyou,about_section_2)

    html = html + logos("about",lang) +"""
    </div>
  </div>
</div>

<div class="white_band" role="complementary">
  <div class="content_bar">
    <div class="content_bar_in">%s

    <div class="content_bar_logos">
        <a href="https://www.mass.gov/orgs/massachusetts-court-system" target="_blank"><img src="%simages/court_seal.png" height="90px" width="90px" alt="%s" class="logo"></a>
    </div>

    </div>
  </div>
</div>"""%(developed,path,court_logo)    
    

    html = html + """</div>
  </div>
</div>"""
    html = html + footer(page,lang)    
    return html

def helpline(lang):
    page="helpline"
    
    help_header = df_static.loc['help_header'][lan[lang]]
    help_body = df_static.loc['help_body'][lan[lang]]
    
    if (lang=="en"):
        suff = ""
    else:
        suff = lang
    path = "../"
    if (lang !="en"):
        path = path + "../"

    html = header(page,lang) + topics(page,lang)        
    html = html + """<div class="white_band">
  <div class="content_bar">
    <div class="content_bar_in_text" style="padding:10px;">
      <center><h1>%s</h1></center>
  </div>
</div>
</div>

<div class="white_band" role="main">
  <div class="content_bar">
    <div class="content_bar_in_text">
      %s
  </div>
</div>
</div>"""%(help_header,help_body)
    html = html + footer(page,lang)    
    return html

def privacy(lang):
    page="privacy"
    
    priv_header = df_static.loc['priv_header'][lan[lang]]
    priv_body = df_static.loc['priv_body'][lan[lang]]
        
    if (lang=="en"):
        suff = ""
    else:
        suff = lang
    path = "../"
    if (lang !="en"):
        path = path + "../"
        
    priv_body = priv_body%(path,suff)
    
    if (lang !="en"):
        priv_body = df_static.loc['priv_trans'][lan[lang]] + priv_body

    html = header(page,lang) + topics(page,lang)
    html = html + """  <div class="white_band">
    <div class="content_bar">
      <div class="content_bar_in_text" style="padding:10px;">
        <h1>%s</h1>
    </div>
  </div>
  </div>

<div class="white_band" role="main">
  <div class="content_bar">
    <div class="content_bar_in_text">
      %s
    </div>
  </div>
</div>"""%(priv_header,priv_body)
    html = html + footer(page,lang)
    
    return html

def myforms(lang):        
    html = """<html>
<head>
<meta http-equiv="refresh" content="0; url=https://apps.suffolklitlab.org/interviews" />
</head>
</html>"""
    
    return html


def beta_209A(lang):
    page="209A beta"
    
    beta_209A_header = df_static.loc['beta_209A_header'][lan[lang]]
    beta_209A_body = df_static.loc['beta_209A_body'][lan[lang]]
        
    if (lang=="en"):
        suff = ""
    else:
        suff = lang
    path = "../"
    if (lang !="en"):
        path = path + "../"
        
    #priv_body = news_body%(path,suff)
    
    html = header(page,lang) + topics(page,lang)
    html = html + """  <div class="white_band">
    <div class="content_bar">
      <div class="content_bar_in_text" style="padding:10px;">
        <center><h1>%s</h1></center>
    </div>
  </div>
  </div>

<div class="white_band" role="main">
  <div class="content_bar">
    <div class="content_bar_in_text">
      %s
    </div>
  </div>
</div>"""%(beta_209A_header,beta_209A_body)
    html = html + footer(page,lang)
    
    return html


def create_listing(page,path,status,lang):
    
    lang_msg = df_static.loc['lang_msg'][lan[lang]]
    start_form = df_static.loc['start'][lan[lang]]
    not_avail = df_static.loc['not_avail'][lan[lang]]
    beta = df_static.loc['beta'][lan[lang]]
    in_prog = df_static.loc['in_prog'][lan[lang]]
    in_progress_msg = df_static.loc['in_progress_msg'][lan[lang]]   

    form_2_base_pdf = df_forms.loc['form_2_base_pdf'][lan[lang]]   
    
    # create an empty dataframe
    df_list = pd.DataFrame([],columns=["html","counts","type"]) 
    
    if (lang=="en"):
        suff = ""
    else:
        suff = lang
    path = "../"
    if (lang !="en"):
        path = path + "../"
        
    for n in range(1,1+math.floor(len(df_forms)/5)):
        
        #print(n)
        
        form_html = ""

        title = df_forms.loc['form_%s_title'%n][lan[lang]]
        anchor = df_forms.loc['form_%s_title'%n]["anchor"] 
        counts = df_forms.loc['form_%s_title'%n]["counts"]
        
        if counts > 0 or counts < 0:
            counts = counts
        else:
            counts = 0
                        
        pdf_name = re.sub(r'[^A-Za-z_]', '', re.sub(r'\s', '_', df_forms.loc['form_%s_title'%n]["English"])) + ".pdf"
        button_desc = df_forms.loc['form_%s_desc'%n][lan[lang]]
        pdf_link = str(df_forms.loc['form_%s_base_pdf'%n]["English"])
        button_onclick = ""

        if (pd.notnull(df_forms.loc['form_%s_link'%n][lan[lang]])):
            if ("beta" in df_forms.loc['form_%s_link'%n][lan[lang]]):
                link_href = path + df_forms.loc['form_%s_link'%n][lan[lang]] + suff
            else:
                link_href = df_forms.loc['form_%s_link'%n][lan[lang]]
            lang_msg = ""
        else:
            link_href = df_forms.loc['form_%s_link'%n]["English"]
            lang_msg = df_static.loc['lang_msg'][lan[lang]]
            #print(lang_msg,button_desc)
        
        if (pd.notnull(title)):
            
            if (pd.notnull(link_href)):
                if ("beta" in link_href):
                    button_style = "beta"
                    button_text = beta
                    button_target = ""
                else: 
                    button_style = "start"
                    button_text = start_form
                    button_target = "target=\"_blank\""
                    
                button_onclick = "onClick=\"gtag('event', 'interview', {'event_category':'start', 'event_label':'%s'});\""%link_href
                button_desc = "<p><b>" + lang_msg + "</b></p>" + str(button_desc)
            else:
                button_style = "in_prog"
                button_target = ""
                link_href = "javascript:alert('%s');"%not_avail            
                button_text = in_prog
                #print(in_progress_msg)
                
                #print(path,pdf_name,pdf_name)
                
                button_desc = "<p><b>" + in_progress_msg%(path,pdf_name,pdf_name) + "</b></p>" + button_desc

            #pd.notnull(df_forms.loc['form_%s_link'%n][lan[lang]]) and
            if (status==0):
                logic_test = pd.notnull(link_href) and ((page.lower() in df_forms.loc['form_%s_categories'%n]["English"].lower()) or ("all" in df_forms.loc['form_%s_categories'%n]["English"].lower()))
                l_type = 1
            elif (status==1):
                logic_test = pd.isnull(link_href) and ((page.lower() in df_forms.loc['form_%s_categories'%n]["English"].lower()) or ("all" in df_forms.loc['form_%s_categories'%n]["English"].lower()))
                l_type = 0

            if (logic_test == True and counts >= 0):
                pdf_good = download_pdf(pdf_link,"%s"%(pdf_name))
                if (pdf_good or pd.notnull(link_href)):
                    form_html = form_html + """<div class="white_band" role="complementary">
      <div class="content_bar">
        <div class="content_bar_description">
          <a name="%s" href="#%s" class="anchor"></a>
          <h2 class="blue">%s</h2>          
          <div class="desc_table">
            <div class="desc_cell">
            <p>
              %s
            </p>
          </div>
          <div class="desc_cell" style="width:1%%">
            <a href="%s" %s %s><div class="%s">%s</div></a>
          </div>
          </div>

        </div>
      </div>
    </div>

    """%(anchor,anchor,title,button_desc,link_href,button_onclick,button_target,button_style,button_text)

                # append a new row
                df_list = df_list.append(pd.DataFrame([[form_html,counts,l_type]], columns=["html","counts","type"]), ignore_index=True)

    #display(df_list)
    
    form_html_output = ""
    for index, row in df_list[df_list["type"]==1].sort_values(by='counts', ascending=False).iterrows():
        form_html_output = form_html_output + row['html']

    for index, row in df_list[df_list["type"]!=1].sort_values(by='counts', ascending=False).iterrows():
        form_html_output = form_html_output + row['html']    
                
    return form_html_output


def topic(page,lang):      

    appeals_header = df_static.loc['appeals_header'][lan[lang]]
    appeals_body = df_static.loc['appeals_body'][lan[lang]]
    help_header = df_static.loc['help_header'][lan[lang]]
    help_body = df_static.loc['help_body'][lan[lang]]
    
    if (lang=="en"):
        suff = ""
    else:
        suff = lang
    path = "../"
    if (lang !="en"):
        path = path + "../"
    
    if (page == "housing"):
        title = df_static.loc['housing'][lan[lang]]
    elif (page == "dv"):
        title = df_static.loc['dv'][lan[lang]]
    elif (page == "guardianship"):
        title = df_static.loc['guardianship'][lan[lang]]
    elif (page == "family"):
        title = df_static.loc['family'][lan[lang]]
    elif (page == "employment"):
        title = df_static.loc['employment'][lan[lang]]
    elif (page == "consumer"):
        title = df_static.loc['consumer'][lan[lang]]
    elif (page == "health"):
        title = df_static.loc['health'][lan[lang]]
    elif (page == "income"):
        title = df_static.loc['income'][lan[lang]]
    elif (page == "immigration"):
        title = df_static.loc['immigration'][lan[lang]]
    elif (page == "education"):
        title = df_static.loc['education'][lan[lang]]
    elif (page == "appeals"):
        title = df_static.loc['appeals'][lan[lang]]
    elif (page == "safety"):
        title = df_static.loc['safety'][lan[lang]]
    elif (page == "other"):
        title = df_static.loc['other'][lan[lang]]
    
    form_html = create_listing(page,path,0,lang)            
    form_html = form_html + create_listing(page,path,1,lang)      
    
    if (page=="appeals"):
        help_html_output = """<div class="white_band" role="complementary">
  <div class="content_bar">
    <div class="content_bar_description">

      <div class="desc_table">
        <div class="desc_cell">
        <h2 class="help_listing">%s</h2>
        <p>
          %s
        </p>
      </div>
      </div>

    </div>
  </div>
</div>

"""%(appeals_header,appeals_body)
    else: 
        help_html_output = ""

        
    help_html_output = help_html_output + """<div class="white_band" role="complementary">
  <div class="content_bar">
    <div class="content_bar_description">

      <div class="desc_table">
        <div class="desc_cell">
        <h2 class="help_listing">%s</h2>
        <p>
          %s
        </p>
      </div>
      </div>

    </div>
  </div>
</div>

"""%(help_header,help_body)

    if (form_html !=""):
        html = header(page,lang) + topics(page,lang)
        html = html + """<div class="white_band">
      <div class="content_bar">
        <div class="content_bar_in_text" style="padding:10px;">
          <center><h1>%s</h1></center>
      </div>
    </div>
    </div>"""%(title)
        html = html + form_html + help_html_output + footer(page,lang)    
        return html
    
    else: 
        html = """<html>
<head>
<meta http-equiv="refresh" content="0; url=%shelpline/%s" />
<meta http-equiv="expires" content="0" />
</head>
</html>"""%(path,suff)
    
    return html


def copy_and_overwrite(from_path, to_path):
    if os.path.exists(to_path):
        shutil.rmtree(to_path)
    shutil.copytree(from_path, to_path)

def write_page(page="home",lang="en"):
    
    if (page=="home"):
        html = home(lang)
        path = ""
    elif (page=="hackathon"):
        html = hackathon(lang)
        path = "hackathon/"
    elif (page=="news"):
        html = news(lang)
        path = "news/"
    elif (page=="about"):
        html = about(lang)
        path = "about/"
    elif (page=="privacy"):
        html = privacy(lang)
        path = "privacy/"
    elif (page=="helpline"):
        html = helpline(lang)
        path = "helpline/"
    elif (page=="myforms"):
        html = myforms(lang)
        path = "myforms/"
    elif (page=="209A beta"):
        html = beta_209A(lang)
        path = "209A_beta/"
    else:
        html = topic(page,lang)
        path = str(page)+"/"
        
    if (lang=="en"):
        lang_path = ""
    else:
        lang_path = lang+"/"
        
    if not os.path.exists("html_output/%s"%(release)):
        os.makedirs("html_output/%s"%(release))
            
    if not os.path.exists("html_output/%s/%s%s"%(release,path,lang_path)):
        os.makedirs("html_output/%s/%s%s"%(release,path,lang_path))
    
    text_file = open("html_output/%s/%s%sindex.html"%(release,path,lang_path), "w")
    text_file.write(html)
    text_file.close()
    
    copy_and_overwrite('core_assets/css', 'html_output/%s/css'%release)
    copy_and_overwrite('core_assets/images', 'html_output/%s/images'%release)
    copy_and_overwrite('core_assets/js_bin', 'html_output/%s/js_bin'%release)
    copy_and_overwrite('core_assets/pdfs', 'html_output/%s/pdfs'%release)

    print("Page (%s) has been written to the website folder."%page)
    
    
```
