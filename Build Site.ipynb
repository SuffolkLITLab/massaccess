```python
def download_pdf(url,name):
    try:
        r = requests.get(url) # create HTTP response object 
        with open("core_assets/pdfs/%s"%name,'wb') as f: 
            f.write(r.content) 
        #print("\nWrote PDF from %s\n"%url)
        return True
    except:
        print("\nERROR WRITING PDF!!! from %s\n"%url)
        return False
```
